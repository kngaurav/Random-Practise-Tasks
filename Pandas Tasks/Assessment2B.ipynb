```python
# list of students who have exceeded expectations i.e codekata score exceeds 15000 

e_expec = df[df["CodeKata Score"] > 15000]
display(e_expec)
e_expec.to_csv('Exceeded_Expectations.csv')
```

```python
# list of students who have reached expectations i.e codekata score between 10000 & 15000

r_expec = df[df["CodeKata Score"].between(10000 , 15000)]
display(r_expec)
r_expec.to_csv('Reached_Expectations.csv')
```

```python
# list of students who have need improvement i.e codekata score between 7000 & 10000

need_improv = df[df["CodeKata Score"].between(7000 , 10000)]
display(need_improv)
need_improv.to_csv('Need_Improvement.csv')
```

###### Department wise codekata performence (pie chart)

```python
# determining all the departments in given data

df["Department"].unique()
```

```python
# calculating department wise average CodeKata scores

CSE = df[df["Department"] == "Computer Science and Engineering"]
a = CSE["CodeKata Score"].mean()

ECE = df[df["Department"] == "Electronics and Communication Engineering"]
b = ECE["CodeKata Score"].mean()

EEE = df[df["Department"] == "Electronics and Electrical Engineering"]
c = EEE["CodeKata Score"].mean()
```

```python
# plotting the pie chart
import matplotlib.pyplot as plt

x = ([a ,b ,c])
mylabels = ["CSE", "ECE", "EEE"]

colors = ['teal', 'magenta', 'gold']
plt.pie(x, labels = mylabels, startangle = 90,autopct='%1.1f%%',radius=1.5, colors=colors)
plt.title('Department wise CodeKata performence')
plt.show() 
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
# determining top 5 performers from each department

cse_toppers = CSE.nlargest(5, ['CodeKata Score'])
ece_toppers = ECE.nlargest(5, ['CodeKata Score'])
eee_toppers = EEE.nlargest(5, ['CodeKata Score'])
```

```python
# CSE toppers

plt.barh(cse_toppers['Name'] , cse_toppers['CodeKata Score'])
plt.title('Computer Science Engineering Toppers')
plt.xlabel('CodeKata Score')
plt.ylabel('Name')
plt.show()
```
